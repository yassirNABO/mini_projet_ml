```python
df = df.dropna()
```

```python
df = df.copy()
```

```python
df['Engine'] = df['Engine'].astype(str).str.extract('(\d+)').astype(float)
df['Max Power'] = df['Max Power'].astype(str).str.extract('(\d+)').astype(float)
df['Max Torque'] = df['Max Torque'].astype(str).str.extract('(\d+)').astype(float)
```

```python
df[['Engine', 'Max Power', 'Max Torque']].head()
df.dtypes
```

```python
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# numeric
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col] = df[col].fillna(df[col].mean())

# categorical
for col in df.select_dtypes(include=['object']):
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Price'] >= Q1 - 1.5*IQR) & (df['Price'] <= Q3 + 1.5*IQR)]
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

X = df.drop('Price', axis=1)
y = df['Price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("LR R2:", r2_score(y_test, y_pred_lr))
print("RF R2:", r2_score(y_test, y_pred_rf))

print("LR MSE:", mean_squared_error(y_test, y_pred_lr))
print("RF MSE:", mean_squared_error(y_test, y_pred_rf))
```
